# Data analysis and visualisation

```python
cancelled_per =df['is_canceled'].value_counts(normalize=True)
cancelled_per
```

```python
plt.figure(figsize=(5,3))
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts(),edgecolor='k')
```

```python
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels=ax1.get_legend_handles_labels()
plt.title('Reservation Status of different hotels')
plt.xlabel('Hotel')
plt.ylabel('number of reservation')
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(5,3))
plt.bar(['Not canceled','canceled'],resort_hotel['is_canceled'].value_counts(),edgecolor='k')
```

```python
resort1_hotel=df[df['hotel']=='City Hotel']
resort1_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=resort1_hotel.groupby('reservation_status_date')[['adr']].mean()

```

```python
plt.figure(figsize=(20,8))
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
plt.title('reservation status per month')
plt.ylabel('no.of reservations')
plt.xlabel('months')
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
```

```python
data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize=30)
sns.barplot(x='month', y='adr', data=data)
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.title("Top 10 country with reservation cancelled")
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
```
